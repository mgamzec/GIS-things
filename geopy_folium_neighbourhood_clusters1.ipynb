```python
import folium
```

```python
map_spore = folium.Map(location=[1.3521,103.8198],zoom_start=10) ## location for Singapore
map_spore
```

```python
map_spore = folium.Map(location=[1.3521,103.8198],zoom_start=10)
for lat,lng,borough,neighbourhood in zip(df['Latitude'],df['Longitude'],df['STN_NAME'],df['COLOR']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_spore)
map_spore
```

```python
latitude = 1.3521
longitude = 103.8198 #singapore
```

```python
DFW_Map = folium.Map(
                location = [latitude, longitude],
                zoom_start = 10)

stations = folium.map.FeatureGroup() #it will let us add different as color-coded children
latitudes = list(df.Latitude)
longitudes = list(df.Longitude)
labels = list(df.COLOR) #list of zones

for lat, lng, label in zip(latitudes, longitudes, labels):
  if label == 'RED':
    folium.Marker(
      location = [lat, lng],
      popup = label,
      icon = folium.Icon(color='red', icon='info-sign')
     ).add_to(DFW_Map)
  else:
    folium.Marker(
      location = [lat, lng],
      popup = label,
      icon = folium.Icon(color='green', icon='info-sign')
     ).add_to(DFW_Map)

DFW_Map.add_child(stations)
DFW_Map
```

Multiple Colors- MRT

```python
DFW_Map = folium.Map(
                location = [latitude, longitude],
                zoom_start = 10)

incidents_accident = folium.map.FeatureGroup()
latitudes = list(df.Latitude)
longitudes = list(df.Longitude)
labels = list(df.COLOR)

for lat, lng, label in zip(latitudes, longitudes, labels):
  if label == 'RED':
    folium.Marker(
      location = [lat, lng],
      popup = label,
      icon = folium.Icon(color='red', icon='info-sign')
     ).add_to(DFW_Map)
  elif label == 'GREEN':
    folium.Marker(
      location = [lat, lng],
      popup = label,
      icon = folium.Icon(color='green', icon='info-sign')
     ).add_to(DFW_Map)
  elif label == 'BLUE':
    folium.Marker(
      location = [lat, lng],
      popup = label,
      icon = folium.Icon(color='blue', icon='info-sign')
     ).add_to(DFW_Map)
  else:
    folium.Marker(
      location = [lat, lng],
      popup = label,
      icon = folium.Icon(color='black', icon='info-sign')
     ).add_to(DFW_Map)

DFW_Map.add_child(incidents_accident)
DFW_Map
```
