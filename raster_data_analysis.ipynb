**Visualization of raster**

```python
from rasterio.plot import show

show(ds)
```

```python
from rasterio.plot import show

show(ds, cmap="Spectral")
```

```python
from rasterio.plot import show

show(ds, cmap="winter", title="Land Cover Map of  Nepal")
```

```python
from rasterio.plot import show, show_hist

show_hist(ds, bins=50, title="Land Cover Map of  Nepal")
```

Writing tiff file

```python
with rasterio.open( r"/content/output/nepal_lc_2020_out.tif", 'w',
                            driver = ds.driver,
                            height = ds.height,
                            width = ds.width,
                            count = ds.count,
                            transform = ds.transform,
                            dtype = data.dtype,
                            )  as dst:

 dst.write(data)
```

```python
ds_out = rasterio.open('/content/output/nepal_lc_2020_out.tif')

show(ds_out)
```
